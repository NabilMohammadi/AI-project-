```python
import seaborn as sns
import pandas as pd   # visar error ,om biblioteket inte har impoterats
df= pd.read_csv("/content/World Happiness Report (2005-2021).csv")
print(df.columns)
```

```python
# visualisera avvikande värden med boxplot
sns.boxplot(data=df, x="Freedom to make life choices")
# Hantera avvikande värden - Exempel på att sätta ett maxvärde
upper_limit = 10  # Exempel: Maxvärdet för 'Freedom to make life choices' ska inte överstiga 10
df.loc[df['Freedom to make life choices'] > upper_limit, 'Freedom to make life choices'] = upper_limit
```

```python
   # tabort kolumner som inte är relevent för analysen
columns_to_drop = ["Generosity" , "Healthy life expectancy at birth"]
df = df.drop(columns=columns_to_drop)
print(df.columns)
```

```python
df = df.drop_duplicates() # ta bort dubbletter från dataframe ,ibland kan den ge fler eller färre rader ,därför ta bort vi dem
```

```python
   # Beräkna tillväxttakt för 'social support'
   # Beräkna tillväxttakt för 'social support'
df['Social support %'] = df['Social support'].pct_change() * 100
 # visa de första rader att se resultat
print(df[["Year", "Social support", "Social support %"]])
```

```python
# Beräkna förändring (diff) för 'Social support'
df['Social support diff'] = df['Social support'].diff()

# Visa de första raderna för att se resultatet
print(df[['Year', 'Social support', 'Social support diff']])
```

```python
# 1. Beräkna proportionen av "Positive affect" i förhållande till "Negative affect",vi delar de med varandra för att hitta portioner
df['Positive to Negative affect'] = df['Positive affect'] / df['Negative affect']
# Visa resultatet
print(df[['Year', 'Positive to Negative affect']])
```

5. Analysera Data i Python:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv("/content/World Happiness Report( med togbort kolumner).csv")
print(df.columns)
```

```python
df = df.drop_duplicates() # ta bort dubbletter från dataframe ,ibland kan den ge fler eller färre rader ,därför ta bort vi dem
```

```python
  # gruppera efter länder och räkna medelvärde Log GDP per capita
df_groupped= df.groupby("Country name")["Log GDP per capita"].mean()
print(df_groupped)
```

```python
# Beräkna korrelationen mellan Life Ladder och GDP per capita
correlation = df[['Life Ladder', 'Log GDP per capita']].corr()
print(correlation)
```

```python
# Linjediagram för att visa Log GDP per capita förändrats över tid
plt.plot(df['Year'], df['Log GDP per capita'])
plt.xlabel('Year')
plt.ylabel('Log GDP per capita')
plt.title('Log GDP per capita over time')
plt.show()
```

```python
# Calculate correlation between variables
correlation_matrix = df[['Life Ladder', 'Social support', 'Log GDP per capita']].corr()

# Skapa en värmekarta (heatmap) för att visa korrelationerna
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap= 'plasma' , fmt='.2f')
plt.title('correlationsmatrix: happiness score, Socialt Support och Log GDP Capita')
plt.show()
```
